1. What are sequence data types in Python?Name them and explain
   indexing and slicing work across different sequence types?

Sequence data types are the data types that store multiple values in
an ordered manner.
Each element has a position (index), allowing access to individual
elements.

Sequence Data Types in Python:
Strings
Lists
Arrays
Tuples
Dictionary
Sets
Range

Indexing:
Indexing is used to access individual elements using their position
number.
Index starts from 0 (left to right).
Negative indexing starts from -1 (right to left).
s = "Python"
print(s[0])   # P
print(s[-1])  # n

Slicing:
Slicing is used to access a group of elements from a sequence.
s = "Python"
print(s[0:3])   # Pyt
print(s[::2])   # Pto


2. What is the difference between a list and a tuple in Python?
   Why would you prefer a tuple over a list in real-world applications?

| Feature    | List                 | Tuple                     |
| ---------- | -------------------- | ------------------------- |
| Symbol     | `[ ]`                | `( )`                     |
| Mutability | Mutable (can change) | Immutable (cannot change) |
| Speed      | Slower               | Faster                    |
| Memory     | Uses more memory     | Uses less memory          |

Why prefer tuple in real-world applications:
When data should not be modified
Faster performance
Uses less memory
Useful for storing fixed data like coordinates, dates, etc.

4. Explain how dictionaries work internally in Python.
   Why are dictionary lookups faster compared to lists or tuples?

A dictionary stores data in key-value pairs.
Example:

d = {"name": "Rahul", "age": 21}

Internal Working:
Dictionaries use a hash table internally.

Process:
Key is passed to a hash function
Hash function generates an index
Value is stored at that index
Retrieval happens directly using the hash value

This concept is called:
👉 Hashing
Why dictionary lookups are faster:
Access time is O(1) (constant time)
No need to search element by element
Direct access using hash value

Compared to lists/tuples:
Lists require sequential search
Time complexity is O(n)


5. What is the difference between sets and lists in Python?
   How do sets handle duplicate values, and where are sets useful
   in practical scenarios?

| Feature    | List    | Set         |
| ---------- | ------- | ----------- |
| Order      | Ordered | Unordered   |
| Duplicates | Allowed | Not allowed |
| Symbol     | `[ ]`   | `{ }`       |
| Indexing   | Yes     | No          |

How sets handle duplicates:
Sets automatically remove duplicate values because each element
must be unique.

Practical Uses of Sets:
Removing duplicates from data
Membership testing (in operator)
Mathematical operations like: Union, Intersection, Difference.
