1. What is a Data Science Pipeline? Explain its key stages with examples.

>A Data Science Pipeline is a systematic workflow used to process raw data and convert it into meaningful insights or predictions. It organizes all steps of data analysis and machine learning into a structured process.

>Key Stages of a Data Science :

* Data Collection  
Raw data is gathered from sources such as CSV files, databases, websites, APIs, or sensors.    
Example : Collecting employee salary data from a company database.  

* Data Cleaning  
Incorrect, missing, or duplicate data is removed or corrected.  
Example: Filling missing salary values using mean or median.  

* Exploratory Data Analysis (EDA)  
Data is analyzed using statistics and graphs to understand patterns and relationships.  
Example: Creating scatter plots to study relationship between experience and salary.  

* Feature Engineering  
New useful features are created from existing data.  
Example: Creating “Salary_Per_Experience” from salary and experience columns.  

* Data Preprocessing  
Data is converted into suitable form using encoding and scaling techniques.  
Example: Converting department names into numerical values.  

* Model Building
Machine learning algorithms are trained using the processed data.  
Example: Using Linear Regression to predict employee salary.  

* Model Evaluation  
Model performance is measured using evaluation metrics.  
Example: Calculating R² Score and RMSE.  

* Deployment    
The trained model is used in real-world applications.  
Example: Deploying salary prediction system on a web application.

3. Differentiate between Label Encoding and OneHot Encoding. When should each be used?

| Label Encoding                            | OneHot Encoding                                   |
| ----------------------------------------- | ------------------------------------------------- |
| Converts categories into numerical labels | Creates separate binary columns for each category |
| Example: HR=0, IT=1, Finance=2            | Example: Separate columns for HR, IT, Finance     |
| Introduces order among categories         | Does not introduce order                          |
| Uses less memory                          | Uses more memory                                  |
| Suitable for ordinal data                 | Suitable for nominal data                         |    

&nbsp;

>When to Use Label Encoding  
Used when categorical values have a meaningful order.   

>Example:  
Low = 0  
Medium = 1  
High = 2  

>When to Use OneHot Encoding
 Used when categories do not have any order.

>Example : Departments like HR, IT, Finance.

4. Why is feature scaling required? Compare normalization and standardization.

>Feature Scaling is used to bring numerical features into a similar range so that no feature dominates others because of larger values.

>It is important for machine learning algorithms such as:  
* KNN  
* SVM  
* Linear Regression  
* Neural Networks  

>Without scaling, features with large values can negatively affect model performance.

>Difference Between Normalization and Standardization -
| Normalization                 | Standardization                                  |
| ----------------------------- | ------------------------------------------------ |
| Scales values between 0 and 1 | Converts data to mean 0 and standard deviation 1 |
| Uses Min-Max Scaling          | Uses Z-score Scaling                             |
| Sensitive to outliers         | Less sensitive to outliers                       |
| Best for bounded data         | Best for normally distributed data               |  
  

>Normalization Formula - X norm  = X - Xmin / Xmax -Xmin

>Standardization Formula : Z = X - μ / σ

>Where:
* μ = mean    
* σ = standard deviation

* Helps capture hidden relationships  
* Makes models more efficient  
* Improves interpretability
