Theory :

Q1. What is the importance of data visualization in data analysis? How does
visualization help in exploratory data analysis (EDA)?

Data visualization is important in data analysis because it helps convert
complex numerical data into visual formats that are easier to understand and
interpret. Instead of examining large tables of numbers, charts and graphs
allow users to quickly observe patterns, relationships, and variations within
the data. Visualization improves communication of insights by presenting
information in a clear and meaningful way that can be easily understood by
both technical and non-technical users.

In exploratory data analysis (EDA), visualization helps analysts explore data
before applying statistical or machine learning techniques. It allows them to
identify trends, detect outliers, understand distributions, and recognize
relationships between variables. By using visual tools such as scatter plots,
histograms, and box plots, analysts can gain initial insights into the data and
make better decisions about cleaning, transforming, and modeling the dataset.


Q2. What is the difference between Matplotlib and other visualization
libraries?

Matplotlib is one of the most widely used visualization libraries in Python
and is known for providing detailed control over plot creation & customization.
It allows users to create a wide variety of basic plots such as line plots,
bar charts, scatter plots, and histograms. Matplotlib is considered the
visualization library because many other libraries are built on top of it or
use it internally for plotting.

Other visualization libraries, such as Seaborn, Plotly, and Bokeh, offer
additional features and simplified syntax for creating more visually
appealing and interactive graphics. For example, Seaborn provides better
default styles and statistical plots, while Plotly and Bokeh support
interactive visualizations that can be used in web applications.
In summary, Matplotlib provides flexibility and control, while other
libraries often focus on improved aesthetics, ease of use, or interactivity.


Q3. What is the purpose of the pyplot module in Matplotlib?

The pyplot module in Matplotlib provides a collection of functions that make
it easy to create and display plots in a simple and interactive manner. It
acts as an interface that allows users to generate figures, add titles,
labels, and legends, and display the final graph with minimal code. Pyplot
follows a procedural style, where each function call adds elements to the
current figure.

Using pyplot, users can quickly create common visualizations such as line
plots, bar charts, scatter plots, and histograms without needing to manage
complex object settings manually. It also provides functions for saving
plots to image files and adjusting layout settings, making it a convenient
and essential module for beginners and professionals working with Matplotlib.


Q4. Explain the process of creating a basic line plot.

Creating a basic line plot involves several simple steps that transform
data into a visual representation. First,the Matplotlib library is imported,
usually along with the pyplot module. Next, the data values to be plotted
are prepared in the form of lists or arrays, where one set of values
represents the horizontal axis (x-axis) and another set represents the
vertical axis (y-axis).

After preparing the data, the plot() function from pyplot is used to draw the
line by connecting the data points. Additional elements such as axis labels,
a title, and a legend can be added to make the graph more informative and
readable. Finally, the show() function is used to display the completed line
plot on the screen. This sequence of steps forms the basic workflow for
creating most visualizations in Matplotlib.


Q5. Why is customization important in data visualization? What customization
options are available in Matplotlib?

Customization is important in data visualization because it improves the
clarity, readability, and visual appeal of graphs. A well-customized chart
makes it easier to understand the data and highlight important information.
Without proper customization, visualizations may appear confusing or fail
to communicate the intended message effectively. Customization also helps
match visualizations to specific presentation or reporting requirements.

Matplotlib provides many customization options that allow users to modify
the appearance of plots according to their needs. These options include
changing colors, line styles, marker types, font sizes, and figure sizes.
Users can also add titles, axis labels, legends, grid lines, and annotations
to improve understanding of the data. Additionally, Matplotlib allows users
to adjust axis limits, background styles, and layout settings to create
clear and professional-looking visualizations.

````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("cleaned_employee_data.csv")

# Display first rows
df.head()
````

````python
# Scatter Plot

plt.figure(figsize=(6,4))

plt.scatter(
    df["Experience"],
    df["Salary"]
)

plt.title("Scatter Plot of Salary vs Experience")
plt.xlabel("Experience (Years)")
plt.ylabel("Salary")

plt.show()
````

````python
# Line Plot

sorted_df = df.sort_values(by="Experience")

plt.figure(figsize=(6,4))

plt.plot(
    sorted_df["Experience"],
    sorted_df["Salary"]
)

plt.title("Line Plot of Salary by Experience")
plt.xlabel("Experience (Years)")
plt.ylabel("Salary")

plt.show()
````

````python
# Bar Plot

dept_salary = df.groupby("Department")["Salary"].mean()

plt.figure(figsize=(6,4))

plt.bar(
    dept_salary.index,
    dept_salary.values
)

plt.title("Average Salary by Department")
plt.xlabel("Department")
plt.ylabel("Average Salary")

plt.xticks(rotation=45)

plt.show()
````

````python
# Histogram

plt.figure(figsize=(6,4))

plt.hist(
    df["Salary"],
    bins=10
)

plt.title("Histogram of Salary Distribution")
plt.xlabel("Salary")
plt.ylabel("Frequency")

plt.show()
````

````python
# Box Plot

plt.figure(figsize=(6,4))

sns.boxplot(
    x="Department",
    y="Salary",
    data=df
)

plt.title("Box Plot of Salary by Department")

plt.xticks(rotation=45)

plt.show()
````

````python
# Pair Plot

sns.pairplot(
    df[[
        "Age",
        "Salary",
        "Experience",
        "Bonus"
    ]]
)

plt.show()
````
