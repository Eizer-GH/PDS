1. What is data ingestion and why is it important in data science?
   Which Python  library is commonly used for data ingestion
   and cleaning?

Ans:
Data ingestion is the process of collecting and importing data from
different sources such as CSV files, databases, APIs, or Excel files
into a system for further processing and analysis.

Data ingestion is important in data science because:
It allows data scientists to gather data from multiple sources.
It prepares raw data for analysis and machine learning.
It ensures data is available in a structured format.
Without proper ingestion, analysis cannot begin.

The most commonly used Python library for data ingestion and
cleaning is Pandas.

Other useful libraries include:
NumPy – for numerical operations
OpenPyXL – for Excel files
CSV module – for reading CSV files


2. Why is data cleaning necessary before performing analysis or
   machine learning?

Ans:
Data cleaning is necessary because real-world data is often
incomplete, inconsistent, or incorrect.

Data cleaning helps to:
Remove missing values
Fix incorrect or duplicate data
Standardize formats
Improve accuracy of results
Enhance performance of machine learning models

If data is not cleaned properly, it can lead to:
Incorrect analysis results
Poor model performance
Misleading conclusions

Clean data acts like polished glass — the clearer it is, the clearer
the insights you see.


3. How do you load a CSV file into a Pandas DataFrame?

Ans:
A CSV file can be loaded into a Pandas DataFrame using the read_csv()
function.

Syntax:

import pandas as pd
df = pd.read_csv("filename.csv")

This command reads the CSV file and stores the data into a DataFrame
named data, which can then be used for analysis.


4. What is the difference between head() and tail() functions
   in Pandas?

Ans:
Both head() and tail() functions are used to view rows of a dataset.

| Function   | Description                                |
| ---------- | ------------------------------------------ |
|   head()   | Displays the first few rows of the dataset |
|   tail()   | Displays the last few rows of the dataset  |

Default Behavior:

head() shows first 5 rows
tail() shows last 5 rows

Example:
df.head(10)
df.tail(10)

These functions are useful for quickly previewing data without
scrolling through the entire dataset.


5. How can you identify missing values in a dataset?

Ans:
Missing values in a dataset can be identified using the isnull()
or isna() functions in Pandas.

Methods to identify missing values:

1] Check missing values in dataset
   df.isnull()

2] Count missing values in each column
   df.isnull().sum()

3] Check if any missing value exists
   df.isnull().any()

These functions help detect incomplete data so that it can be
cleaned or filled before performing analysis.

````python
import pandas as pd
import numpy as np
````

````python
df = pd.read_csv("/content/employee_data.csv")  #Loads CSV file
````

````python
df.head()
````

````python
# Check missing values
df.isnull().sum()
````

````python
# Replace missing values with zero
df_filled = df.fillna(0)
df_filled.head(20)
````

````python
# Print column names vertically in lowercase
df.columns = df.columns.str.lower()
for col in df.columns:
    print(col)
````

````python
df.shape
````

````python
df = df.drop_duplicates()
````

````python
# Fill numeric missing values using mean
df.fillna(df.mean(numeric_only=True), inplace=True)
df.isnull().sum()
df.head(20)
````

````python
# Convert selected columns from float to int
df['age'] = df['age'].astype(int)
df['salary'] = df['salary'].astype(int)
df['experience'] = df['experience'].astype(int)
df['performance_rating'] = df['performance_rating'].astype(int)
df['bonus'] = df['bonus'].astype(int)
df['work_hours_per_week'] = df['work_hours_per_week'].astype(int)

df.head()
````

````python
# Filter employees with salary > 100000
filtered_df = df[df['salary'] > 100000]
filtered_df.iloc[:, :7].head()
````

````python
# Save cleaned dataset
df.to_csv("cleaned_employee_data.csv", index=False)
````
