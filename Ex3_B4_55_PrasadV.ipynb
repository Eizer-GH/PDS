Q1. What is data wrangling, and why is it important before analysis?
Answer:
Data wrangling is the process of cleaning, transforming, and organizing raw data into
a structured format suitable for analysis. It includes handling missing values,
correcting errors, removing duplicates, and converting data types.

Data wrangling is important before analysis because raw data is often incomplete,
inconsistent, or incorrectly formatted. Cleaning the data ensures accuracy, improves
data quality, and helps produce reliable results during analysis and visualization.


Q2. You receive messy data with missing values, duplicates, and wrong types.
What steps will you follow to clean it?
Answer:
The following steps can be followed to clean messy data in a systematic and efficient
manner before starting any form of analysis.

1. Load the dataset using pandas so that the data becomes available for inspection
   and further processing.
2. Check the structure of the dataset using functions like head(), info(), and
   describe() to understand the layout and condition of the data.
3. Identify missing values using isnull() or notnull() to locate incomplete entries.
4. Handle missing values by removing unnecessary rows or columns, or filling them
   with suitable values using fillna().
5. Identify duplicate rows using duplicated() and remove them using
   drop_duplicates() to avoid repeated records.
6. Check data types using dtypes and convert incorrect types using astype()
   so that calculations can be performed correctly.
7. Rename incorrect or unclear column names if required to improve readability
   and understanding of the dataset.
8. Detect and handle outliers if necessary to prevent extreme values from affecting
   the analysis results.
9. Verify the cleaned dataset carefully before performing further analysis.


Q3. How can you check column names, data types, and summary statistics?
What does info() show that describe() does not?
Answer:
Column names can be checked using df.columns to display all available column labels.
Data types can be checked using df.dtypes to understand the format of stored values.
Summary statistics can be checked using df.describe() to view statistical measures.

The info() function provides details about column names, number of non-null values,
data types of each column, and memory usage of the dataset, which helps understand
the structure of the data.

The describe() function provides statistical summaries such as mean, median,
standard deviation, minimum and maximum values, and percentiles, which help in
understanding the distribution of numerical data.

The info() function shows non-null counts and memory usage, which describe() does
not provide, making it useful for checking missing data and memory optimization.


Q4. How do you identify missing values and outliers in a dataset?
Answer:
Missing values can be identified using df.isnull() and df.isnull().sum(), which help
detect and count missing entries in each column of the dataset.

Outliers can be identified using summary statistics from describe(), visualization
methods such as box plots and scatter plots, identifying values that are far from
the average, and statistical methods like Interquartile Range (IQR) or Z-score.

Outliers are values that are significantly higher or lower than the rest of the
data and may affect the accuracy of analysis if not handled properly.


Q5. How do you reduce memory usage in pandas DataFrames?
Answer:
Memory usage in pandas DataFrames can be reduced by converting data types to smaller
types such as using int32 instead of int64, which reduces the amount of storage used
by numerical values.

Another useful method is converting object data types to category when possible,
as categorical data requires less memory compared to object data types.

Removing unnecessary columns that are not needed for analysis also helps reduce
memory usage and improves processing speed.

Using efficient file formats or compressed formats while saving data helps reduce
storage size, and checking memory usage using df.memory_usage() helps identify
large columns that need optimization.

````python
import pandas as pd

# Load dataset
df = pd.read_csv("cleaned_employee_data.csv")

# Display first few rows
df.head()
````

````python
# Select single column (Salary)
salary_column = df["Salary"]

salary_column.head()
````

````python
# Add new column with constant value
df["Company_Name"] = "ABC_Corp"
df.iloc[:, :7].head()
df.iloc[:, 7:].head()
````

````python
# Sort DataFrame by Age
sorted_df = df.sort_values(by="Age")

sorted_df.iloc[:, :4].head()
````

````python
# Rename column
df = df.rename(columns={
    "Work_Hours_Per_Week": "Weekly_Hours"
})

df.head()
````

````python
# Summary statistics
df.describe()
````

````python
# Filter employees with Salary > 70000 and Experience < 7
filtered_df = df[
    (df["Salary"] > 70000) &
    (df["Experience"] < 7)
]

filtered_df.iloc[:, :8].head()
````

````python
# Group by Department and calculate mean Salary
grouped_df = df.groupby("Department")["Salary"].mean()

grouped_df
````

````python
# Create small second DataFrame
dept_info = pd.DataFrame({
    "Department": ["HR", "Marketing", "IT"],
    "Location": ["Delhi", "Mumbai", "Pune"]
})

# Merge with original DataFrame
merged_df = pd.merge(df, dept_info, on="Department")

merged_df.iloc[:, 4:][5:13]
````

````python
# Drop columns
df_dropped = df.drop(columns=["Company_Name","Performance_Rating"])

# Drop rows with index 0 and 1
df_dropped = df_dropped.drop(index=[0, 1])

df_dropped.head()
````

````python
# Increase bonus by 10%

def increase_Bonus(x):
    return x * 1.10

df["Updated_Bonus"] = df["Bonus"].apply(increase_Bonus)

df.iloc[:, 9:].head()
````

````python
# Task 11A: Pivot Operation

pivot_table = df.pivot_table(
    values="Age",
    index="Department",
    aggfunc="mean"
)

pivot_table
````

````python
# Task 11B: Melt Operation

melted_df = pd.melt(
    df,
    id_vars=["Employee_ID", "Department"],
    value_vars=["Salary", "Bonus"],
    var_name="Pay_Type",
    value_name="Amount"
)

melted_df.head()
````
