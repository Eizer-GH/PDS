Theory :

Q1. What is web scraping? Explain the role of the Requests library
in web scraping.
Answer:
Web scraping is the process of automatically extracting data from websites
using programs or scripts. It allows users to collect useful information
such as text, images, tables, or links from web pages instead of copying
data manually.

The Requests library in Python is used to send HTTP requests to websites. It
helps in downloading the HTML content of a web page so that it can be analyzed
and processed. In web scraping, Requests is used to:
• Send a request to a website
• Receive the webpage content
• Access the HTML source code
• Pass the HTML data to BeautifulSoup for parsing


Q2. What is BeautifulSoup? How does it help in parsing HTML?
Answer:
BeautifulSoup is a Python library used for parsing HTML and XML
documents. It helps in extracting specific information from web
pages in an easy and structured way.

BeautifulSoup helps in parsing HTML by:
• Converting raw HTML into a tree structure
• Allowing users to search for specific tags and elements
• Extracting text, links, images, and other data
• Navigating through HTML elements using tags and attributes
• Making it easier to clean and organize web data


Q3. Differentiate between web scraping and web crawling.
Answer:
Web Scraping:
Web scraping is the process of extracting specific data from websites.
It focuses on collecting useful information from selected web pages.

Web Crawling:
Web crawling is the process of automatically browsing and indexing
multiple web pages across the internet. It focuses on discovering
new pages and links.

Difference:
• Web scraping extracts specific data from web pages.
• Web crawling searches and indexes many web pages.
• Scraping focuses on data collection.
• Crawling focuses on navigation and page discovery.


Q4. What are the commonly used methods in BeautifulSoup?
Answer:
Some commonly used methods in BeautifulSoup include:

• find() – Finds the first matching HTML element.
• find_all() – Finds all matching HTML elements.
• get_text() – Extracts text from HTML tags.
• select() – Selects elements using CSS selectors.
• select_one() – Selects the first element matching a CSS selector.
• prettify() – Formats HTML content in a readable structure.
• get() – Retrieves the value of an attribute from a tag.


Q5. What are the ethical and legal considerations in web scraping?

Answer:
Web scraping should be performed responsibly by following ethical
and legal rules. Important considerations include:

• Respect the website's robots.txt file rules.
• Do not overload servers with too many requests.
• Avoid scraping private or restricted data.
• Follow copyright laws and website terms of service.
• Use scraped data only for legal and ethical purposes.
• Give proper credit when using data from websites.

````python
# Importing required libraries
# requests → used to send HTTP requests to a website
# bs4 (BeautifulSoup) → used to parse HTML content
import requests
from bs4 import BeautifulSoup

# Step 1: Define the URL of the webpage you want to scrape
# Example: A simple static website for practice
url = "https://quotes.toscrape.com/"

# Step 2: Send an HTTP GET request to the URL
# This fetches the webpage content
response = requests.get(url)

# Step 3: Check if the request was successful
# Status code 200 means OK (successful request)
if response.status_code == 200:
    print("Successfully fetched the webpage")
else:
    print("Failed to fetch webpage")
    exit()  # Stop execution if request fails

# Step 4: Parse the HTML content using BeautifulSoup
# 'html.parser' is Python's built-in HTML parser
soup = BeautifulSoup(response.text, 'html.parser')

# Step 5: Find all quote containers on the page
# Each quote is inside a <div> with class 'quote'
quotes = soup.find_all('div', class_='quote')

# Step 6: Loop through each quote and extract data
for index, quote in enumerate(quotes):

    # Extract the quote text (inside <span class="text">)
    text = quote.find('span', class_='text').get_text()

    # Extract the author name (inside <small class="author">)
    author = quote.find('small', class_='author').get_text()

    # Extract tags (inside <a class="tag">)
    tags = quote.find_all('a', class_='tag')

    # Convert list of tag elements into actual text list
    tag_list = [tag.get_text() for tag in tags]

    # Print extracted data
    print(f"\nQuote {index + 1}:")
    print("Text:", text)
    print("Author:", author)
    print("Tags:", ", ".join(tag_list))

# Step 7: (Optional) Extract page title
title = soup.find('title').get_text()
print("\nPage Title:", title)
````

````python
# Import required libraries
import requests                     # For sending HTTP requests
from bs4 import BeautifulSoup      # For parsing HTML

# Step 1: Take URL input from user
url = input("Enter the URL to scrape: ")

# Step 2: Take HTML tag input (e.g., div, p, h1, a)
tag = input("Enter the HTML tag you want to extract (e.g., div, p, h1, a): ")

# Step 3: Optional - take class name input
# User can leave blank if not needed
class_name = input("Enter class name (leave blank if not required): ")

# Step 4: Send HTTP GET request to the URL
try:
    response = requests.get(url)

    # Step 5: Check response status
    if response.status_code != 200:
        print("Failed to fetch webpage. Status code:", response.status_code)
        exit()
except Exception as e:
    print("Error while requesting webpage:", e)
    exit()

# Step 6: Parse HTML content using BeautifulSoup
soup = BeautifulSoup(response.text, 'html.parser')

# Step 7: Extract elements based on user input
# If class name is provided → use it
if class_name:
    elements = soup.find_all(tag, class_=class_name)
else:
    elements = soup.find_all(tag)

# Step 8: Check if elements found
if not elements:
    print("No elements found with given criteria.")
else:
    print(f"\nFound {len(elements)} elements:\n")

# Step 9: Loop through elements and print text + attributes
for index, element in enumerate(elements, start=1):

    print(f"Element {index}:")

    # Extract and print text inside the tag
    text = element.get_text(strip=True)
    print("Text:", text if text else "No text available")

    # Extract and print all attributes (like href, src, etc.)
    print("Attributes:", element.attrs)

    print("-" * 50)

# Step 10: Extract page title
title = soup.title.string if soup.title else "No title found"
print("\nPage Title:", title)
````
