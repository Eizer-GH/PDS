Theory -

1. Difference between Binomial, Poisson, and Normal distributions     

| Feature    | Binomial Distribution       | Poisson Distribution       | Normal Distribution         |
| ---------- | --------------------------- | -------------------------- | --------------------------- |
| Type       | Discrete                    | Discrete                   | Continuous                  |
| Used for   | Fixed number of trials      | Events in a fixed interval | Natural continuous data     |
| Parameters | n (trials), p (probability) | λ (rate of occurrence)     | μ (mean), σ (std deviation) |
| Example    | Tossing a coin n times      | Number of calls per hour   | Heights, marks, weights     |
| Shape      | Can be skewed or symmetric  | Positively skewed          | Symmetrical bell curve      |

2. What is the real-world significance of the Poisson distribution?

>Poisson distribution is used when we count how many times an event happens in a fixed interval.  

>Examples:  
>* Number of calls in a call center per hour   
>* Number of accidents on a road per day  
>* Number of emails received per minute  
>* Website traffic (requests per second)  

>Why important?  

>Because it helps in:  
>* Predicting rare events  
>* Resource planning (staffing, servers)  
>* Queue management systems

3. What is the Central Limit Theorem and how does it relate to Normal distribution?

>The Central Limit Theorem (CLT) states:

> x̄ ∼ N ( μ, σ​ / √n​)

>Meaning:  
When you take large samples from any distribution, The sampling mean tends to follow a Normal distribution,   
Even if the original data is not normal.  

>Relation to Normal Distribution : CLT is the reason why Normal distribution appears everywhere  

>It justifies using Normal distribution in:  

>* Statistics  
>* Machine learning  
>* Data analysis  

&nbsp;

4. Why is Normal distribution widely used in machine learning and statistics?  

>Because:  
>* Many real-world data naturally follow it (marks, heights, errors)  
>* It is mathematically simple and well understood  
>* Supported by CLT (works even if data isn’t perfectly normal)  

>In Machine Learning:  
>* Used in Gaussian models  
>* Assumption in many algorithms (like Naive Bayes)  
>* Helps in error analysis and prediction  

>Key advantage : It makes calculations easier and models more stable.

5. What is skewness in distributions and which distributions exhibit it?

>What is Skewness ? - Skewness tells us how asymmetric a distribution is.

>* Zero skewness → perfectly symmetric  
>* Positive skew → tail on right side  
>* Negative skew → tail on left side   

>Which distributions show skewness?  
>* Binomial Distribution - Can be skewed depending on probability (p ≠ 0.5)
>* Poisson Distribution - Usually positively skewed, especially when λ is small
>* Normal Distribution - No skewness (perfectly symmetric)


```python


import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom, poisson, norm

# Set up a figure with subplots
fig, axs = plt.subplots(3, 3, figsize=(15, 12))
fig.suptitle("Analysis of Binomial, Poisson, and Normal Distributions", fontsize=16)

# ------------------- BINOMIAL DISTRIBUTION -------------------
n, p = 10, 0.5
x_binom = np.arange(0, n+1)
pmf_binom = binom.pmf(x_binom, n, p)
cdf_binom = binom.cdf(x_binom, n, p)
samples_binom = binom.rvs(n, p, size=1000)

# PMF
axs[0, 0].bar(x_binom, pmf_binom, color='skyblue')
axs[0, 0].set_title("Binomial PMF (n=10, p=0.5)")
# CDF
axs[0, 1].step(x_binom, cdf_binom, where='mid', color='blue')
axs[0, 1].set_title("Binomial CDF")
# Histogram of Samples
axs[0, 2].hist(samples_binom, bins=np.arange(n+2)-0.5, density=True, color='navy', alpha=0.7)
axs[0, 2].set_title("Binomial Samples Histogram")

# ------------------- POISSON DISTRIBUTION -------------------
mu = 4
x_poisson = np.arange(0, 15)
pmf_poisson = poisson.pmf(x_poisson, mu)
cdf_poisson = poisson.cdf(x_poisson, mu)
samples_poisson = poisson.rvs(mu, size=1000)

# PMF
axs[1, 0].bar(x_poisson, pmf_poisson, color='lightgreen')
axs[1, 0].set_title("Poisson PMF (μ=4)")
# CDF
axs[1, 1].step(x_poisson, cdf_poisson, where='mid', color='green')
axs[1, 1].set_title("Poisson CDF")
# Histogram of Samples
axs[1, 2].hist(samples_poisson, bins=np.arange(16)-0.5, density=True, color='darkgreen', alpha=0.7)
axs[1, 2].set_title("Poisson Samples Histogram")

# ------------------- NORMAL DISTRIBUTION -------------------
mu_norm, sigma = 0, 1
x_norm = np.linspace(-4, 4, 1000)
pdf_norm = norm.pdf(x_norm, mu_norm, sigma)
cdf_norm = norm.cdf(x_norm, mu_norm, sigma)
samples_norm = norm.rvs(mu_norm, sigma, size=1000)

# PDF
axs[2, 0].plot(x_norm, pdf_norm, color='salmon')
axs[2, 0].set_title("Normal PDF (μ=0, σ=1)")
# CDF
axs[2, 1].plot(x_norm, cdf_norm, color='red')
axs[2, 1].set_title("Normal CDF")
# Histogram of Samples
axs[2, 2].hist(samples_norm, bins=30, density=True, color='darkred', alpha=0.7)
axs[2, 2].set_title("Normal Samples Histogram")

# Layout and show
for row in axs:
    for ax in row:
        ax.grid(True)
plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()

# ------------------- BASIC STATISTICAL ANALYSIS -------------------
print("Binomial Distribution:")
print(f"Mean = {np.mean(samples_binom):.2f}, Variance = {np.var(samples_binom):.2f}")
print("\nPoisson Distribution:")
print(f"Mean = {np.mean(samples_poisson):.2f}, Variance = {np.var(samples_poisson):.2f}")
print("\nNormal Distribution:")
print(f"Mean = {np.mean(samples_norm):.2f}, Standard Deviation = {np.std(samples_norm):.2f}")
```
